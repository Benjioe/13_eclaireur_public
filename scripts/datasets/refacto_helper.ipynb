```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets_file_folder = Path("../../data/datasets/")
refacto_filtered_datasets_filename = 'filtered_datasets.csv'

dataset_scripts_folder = Path ("../../scripts/datasets/")
original_filtered_datasets_filename = 'dataset_filtered_subventions.csv'

refacto_filtered_datasets_df = pd.read_csv(datasets_file_folder / refacto_filtered_datasets_filename)
original_filtered_datasets_df = pd.read_csv(dataset_scripts_folder / original_filtered_datasets_filename)
```

```python
print("Forme de refacto_filtered_datasets_df:", refacto_filtered_datasets_df.shape)
print("Forme de original_filtered_datasets_df:", original_filtered_datasets_df.shape)
```

```python
original_filtered_datasets_df = original_filtered_datasets_df[refacto_filtered_datasets_df.columns]
merged_df = refacto_filtered_datasets_df.reset_index().merge(original_filtered_datasets_df.reset_index(), how='outer', indicator=True, suffixes=('_refacto', '_original'))
difference_df = merged_df[merged_df['_merge'] == 'both']
print("Lignes avec des différences:")
print(difference_df)
```

```python
#merged_df = refacto_filtered_datasets_df.reset_index().merge(original_filtered_datasets_df.reset_index(), how='outer', indicator=True, suffixes=('_refacto', '_original'))
merged_df = refacto_filtered_datasets_df.merge(original_filtered_datasets_df, how='outer', on='id', indicator=True, suffixes=('_refacto', '_original'))
missing_in_refacto = merged_df[merged_df['_merge'] == 'right_only']
difference_df = merged_df[merged_df['_merge'] == 'both']
print("Lignes avec des différences:")
print(difference_df)
```

```python
common_columns = [col for col in original_filtered_datasets_df.columns if col != 'id']

for idx, row in difference_df.iterrows():
    print(f"Ligne {idx}:")
    for col in common_columns:
        if row[col + '_refacto'] != row[col + '_original']:
            print(f"  Colonne {col}: {row[col + '_refacto']} != {row[col + '_original']}")
    print()
```

```python
missing_in_refacto = merged_df[merged_df['_merge'] == 'right_only']
missing_in_original = merged_df[merged_df['_merge'] == 'left_only']

print("Lignes manquantes dans refacto_filtered_datasets_df:")
print(missing_in_refacto)

print("\nLignes manquantes dans original_filtered_datasets_df:")
print(missing_in_original)
```
