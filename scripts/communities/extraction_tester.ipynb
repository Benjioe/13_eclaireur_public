```python
import pandas as pd
```

```python
regions = pd.read_csv("../../data/communities/processed_data/identifiants_regions_2023.csv")
departements = pd.read_csv("../../data/communities/processed_data/identifiants_departements_2023.csv")
communes = pd.read_csv("../../data/communities/processed_data/identifiants_communes_2023.csv")
interco = pd.read_csv("../../data/communities/processed_data/identifiants_epci_2023.csv")
infos_collectivites = pd.read_csv("../../data/communities/processed_data/infos_collectivites.csv")
```

```python
print("Infos Collectivites:")
print(infos_collectivites.describe(include='all'))

print("Nombre de valeurs manquantes par colonne :")
print(infos_collectivites.isnull().sum())

print("Nombre de lignes dupliquées :")
print(infos_collectivites.duplicated().sum())

print("Types uniques de collectivités :")
print(infos_collectivites['type'].unique())
```

```python
print("Exemple de lignes pour chaque type de collectivité :")
for t in infos_collectivites['type'].unique():
    print(f"\nType : {t}")
    print(infos_collectivites[infos_collectivites['type'] == t].head())
```

```python
unique_types = infos_collectivites["type"].unique()

for collectivity_type in unique_types:
    print(f"Type: {collectivity_type}")
    display(infos_collectivites[infos_collectivites["type"] == collectivity_type].head(3))
    print("\n")
```

```python
epci_communes = pd.read_excel("../../data/communities/scrapped_data/gouv_colloc/epcicom2023.xlsx", dtype={"siren": str, "siren_membre": str})
```

```python
epci_communes_grouped = epci_communes.groupby('siren_membre')['siren'].count().reset_index()
epci_communes_grouped.columns = ['siren_membre', 'epci_count']
```

```python
multiple_epci = epci_communes_grouped[epci_communes_grouped['epci_count'] > 1]
print("Nombre de communes avec plusieurs EPCI :", len(multiple_epci))
```
